```python
!pip install folium
import pandas as pd
import folium
from branca.element import Template, MacroElement

# Sample data
data = {
    'City': ['Paris', 'London', 'Dubai', 'Bangkok', 'Singapore',
             'New York', 'Istanbul', 'Kuala Lumpur', 'Tokyo', 'Seoul',
             'Hong Kong', 'Mumbai', 'Rome', 'Osaka', 'Taipei',
             'Mecca', 'Pattaya', 'Phuket', 'Miami', 'Barcelona'],
    'Latitude': [48.8566, 51.5074, 25.2769, 13.7563, 1.3521,
                 40.7128, 41.0082, 3.1390, 35.6762, 37.5665,
                 22.3193, 19.0760, 41.9028, 34.6937, 25.0330,
                 21.3891, 12.9236, 7.9519, 25.7617, 41.3851],
    'Longitude': [2.3522, -0.1278, 55.2962, 100.5018, 103.8198,
                  -74.0060, 28.9784, 101.6869, 139.6503, 126.9780,
                  114.1694, 72.8777, 12.4964, 135.5023, 121.5654,
                  39.8579, 100.8825, 98.3381, -80.1918, 2.1734],
    'Rank': list(range(1, 21)),
    'Visitors (Millions)': [85.2, 80.2, 78.4, 73.8, 70.4,
                            68.2, 65.7, 63.2, 60.3, 58.4,
                            56.3, 54.8, 52.5, 50.9, 49.1,
                            47.8, 45.6, 43.2, 41.5, 39.8]
}

df = pd.DataFrame(data)

# Create color mapping function
def get_color(rank):
    if rank <= 5:
        return 'red'
    elif 6 <= rank <= 10:
        return 'blue'
    elif 11 <= rank <= 15:
        return 'green'
    else:
        return 'purple'

# Create base map
m = folium.Map(location=[20, 0], zoom_start=2, tiles='cartodbpositron')

# Add markers with colors
for index, row in df.iterrows():
    popup_text = f"""
    <b>{row['City']}</b><br>
    Rank: {row['Rank']}<br>
    Visitors: {row['Visitors (Millions)']} Million
    """

    folium.CircleMarker(
        location=[row['Latitude'], row['Longitude']],
        radius=7,
        color=get_color(row['Rank']),
        fill=True,
        fill_color=get_color(row['Rank']),
        popup=folium.Popup(popup_text, max_width=250),
        fill_opacity=0.7
    ).add_to(m)

# Create legend HTML
legend_html = '''
     <div style="position: fixed;
                 bottom: 50px; right: 50px; width: 160px; height: 130px;
                 border:2px solid grey; z-index:9999; font-size:14px;
                 background-color:white;
                 ">
       <div style="padding: 5px">
         <b>Rank Categories</b><br>
         <i style="background:red; width:20px; height:20px; display:inline-block;"></i> 1-5<br>
         <i style="background:blue; width:20px; height:20px; display:inline-block;"></i> 6-10<br>
         <i style="background:green; width:20px; height:20px; display:inline-block;"></i> 11-15<br>
         <i style="background:purple; width:20px; height:20px; display:inline-block;"></i> 16-20
       </div>
     </div>
'''

# Add legend to map
m.get_root().html.add_child(folium.Element(legend_html))

# Display map
m.save('map.html')
m
```
